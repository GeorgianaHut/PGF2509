```python
#oefening 5

def bereken_gemiddelde(getallen):
    if len(getallen) == 0:
        return 0  
    return sum(getallen) / len(getallen)

# Voorbeeld:
lijst = [10, 20, 30, 40]
print(bereken_gemiddelde(lijst))
```

```python
#Oefening 1: Temperature Converter
def fahrenheit_naar_celsius(f):
    return (f - 32) * 5/9

# Testcases
print(fahrenheit_naar_celsius(32))   
print(fahrenheit_naar_celsius(68))   
print(fahrenheit_naar_celsius(100))  
```

```python
# Oefening 3: FizzBuzz

for i in range(1, 21):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
#Oefening 4: Cijfers:
# Dictionary met studenten en hun cijfers
studenten = {
    "Anna": [8, 7, 9],
    "Bram": [6, 5, 7],
    "Celine": [9, 9, 8]
}

# Rapport printen
for naam, cijfers in studenten.items():
    gemiddelde = sum(cijfers) / len(cijfers)
    print(f"{naam} heeft een gemiddeld cijfer van {gemiddelde:.1f}")
```
